### Question 12 [6 pts]

**Code Task A — Vanilla RNN: Shapes and Parameter Count**

Complete the code below so that it:
1. Creates a dummy input tensor matching the setup in Question 11.
2. Initialises a vanilla `nn.RNN` with `hidden_size = 128`.
3. Prints the shapes of `rnn_output` and `h_n`.
4. Prints the total number of parameters in the RNN layer.

```python
import torch
import torch.nn as nn

# --- Setup (from Questions 2 and 11) ---
batch_size  = 32
seq_length  = 15
input_size  = 100
hidden_size = 128

# Task A1: Create the input tensor
x = torch.randn(batch_size, seq_length, input_size)

print(f"Input shape : {x.shape}")

# Task A2: Initialise an RNN layer with the said input_size and hidden_size and set batch_first=True
rnn = nn.RNN(input_size, hidden_size, batch_first=True)

# Task A3: Run a forward pass
rnn_output, h_n = rnn(x)

print(f"rnn_output shape : {rnn_output.shape}")
print(f"h_n shape        : {h_n.shape}")

# Count parameters
total_params = sum(p.numel() for p in rnn.parameters())
print(f"Total RNN parameters : {total_params}")
# Does this match your answer from Question 2B in the worksheet? -- Yes!
```

### Question 13 [6 pts]

**Code Task B — LSTM: Two States and Parameter Count**

Reuse the same `x` tensor from Task A. Complete the code so that it:
1. Initialises an `nn.LSTM` with the same `input_size` and `hidden_size`.
2. Runs a forward pass and prints shapes of outputs.
3. Prints parameter count and compares with RNN.

```python
# Task B1: Initialise an LSTM
lstm = nn.LSTM(input_size, hidden_size, batch_first=True)

# Task B2: LSTM forward pass
lstm_output, (lstm_h_n, lstm_c_n) = lstm(x)

print(f"lstm_output shape : {lstm_output.shape}")
print(f"lstm_h_n shape    : {lstm_h_n.shape}")
print(f"lstm_c_n shape    : {lstm_c_n.shape}")

# Task B3: Parameter count comparison
lstm_params = sum(p.numel() for p in lstm.parameters())
print(f"LSTM params : {lstm_params}")
print(f"Ratio LSTM / RNN : {lstm_params / total_params:.1f}x")
# Does this match your answer from Question 8 in the worksheet? -- Yes!
```
