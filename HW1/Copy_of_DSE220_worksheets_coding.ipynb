## Question 4: Simulating the Learning Rate

Consider the one-dimensional loss function $L(w) = (w - 4)^2$.  
The true minimum is at $w^* = 4$.

Complete the function below to simulate gradient descent for `iterations` steps starting at `w_start`.  
Then run it with:
- `alpha = 0.1` — should converge smoothly toward $w^* = 4$
- `alpha = 1.5` — observe what happens and explain why

**Hint:** The gradient of $L(w) = (w-4)^2$ with respect to $w$ is $\frac{dL}{dw} = 2(w - 4)$.

```python
def gradient_descent(w_start, alpha, iterations):
    w = w_start
    for i in range(iterations):
        # 1. Compute the gradient of L(w) = (w - 4)^2
        #    Hint: dL/dw = 2 * (w - 4)
        grad = 2*(w-4)


        # 2. Update the weight using the gradient descent rule: w = w - alpha * grad
        w    = w - alpha * grad



        print(f"Step {i+1}: w = {w:.4f}")
    return w
```

```python
# Run with a small learning rate — should converge toward w* = 4
print("=== alpha = 0.1 ===")
gradient_descent(w_start=0, alpha=0.1, iterations=5)
```

```python
# Run with a large learning rate — what happens?
print("=== alpha = 1.5 ===")
gradient_descent(w_start=0, alpha=1.5, iterations=5)
```

**Discussion:** What happens when `alpha = 1.5`? Why?

When alpha = 1.5 the model diverges. Each step overshoots the min and lands on the other side, getting worse every iteration instead of converging. This shows why picking the right learning rate is important.

## Question 2: Gradient Descent for Logistic Regression

Perform **one gradient descent update** for a logistic regression model.

**Setup:**
- Features: $x_1 = 2,\; x_2 = 1$, true label $t = 1$
- Initial parameters: $w_1 = 0.5,\; w_2 = 0.5,\; b = 0$
- Learning rate: $\alpha = 0.1$

**Hints:**
- Sigmoid: $\sigma(z) = \dfrac{1}{1 + e^{-z}}$
- Pre-activation: $a = w_1 x_1 + w_2 x_2 + b$
- Prediction: $y = \sigma(a)$
- Error signal (delta): $\delta = t - y$
- Gradients: $\nabla_{w_i} = -\delta \cdot x_i$ (i.e., gradient of loss $= -\delta \cdot x_i$)
- Update rule: $w \leftarrow w - \alpha \cdot \text{grad}$

```python
import math

x1, x2, t = 2.0, 1.0, 1.0
w1, w2, b  = 0.5, 0.5, 0.0
alpha      = 0.1

def sigmoid(z):
    # Hint: return 1 / (1 + e^{-z})
    return 1 / (1 + math.exp(-z))
    pass

# 1. Pre-activation: a = w1*x1 + w2*x2 + b
a = w1*x1 + w2*x2 + b

# 2. Prediction: y = sigmoid(a)
y = sigmoid(a)

# 3. Error signal (delta): delta = t - y
delta = t - y

# 4. Gradients of the loss w.r.t. each parameter
#    grad_w = -delta * x  (negative because loss = -log-likelihood)
grad_w1 = -delta * x1
grad_w2 = -delta * x2
grad_b  =   -delta * 1   # Hint: bias has input 1

# 5. Update weights: w_new = w - alpha * grad
w1_new = w1 - alpha * grad_w1
w2_new = w2 - alpha * grad_w2
b_new  = b  - alpha * grad_b

print(f"Pre-activation a = {a:.4f}")
print(f"Prediction     y = {y:.4f}")
print(f"Delta (t - y)    = {delta:.4f}")
print(f"Updated weights: w1={w1_new:.4f}, w2={w2_new:.4f}, b={b_new:.4f}")
```

## Question 3: The Softmax Function

For multi-class classification, the **softmax function** converts raw logits into a valid probability distribution.

Given $\mathbf{a} = [2, 1, 0]$ for 3 classes, compute the softmax probabilities.

**Hint:** $\text{softmax}(a_k) = \dfrac{e^{a_k}}{\sum_j e^{a_j}}$

**Numerical stability tip:** Subtract `max(a)` from all logits before exponentiating — this prevents overflow and does not change the result since the constant cancels in numerator and denominator.

```python
import numpy as np

logits = np.array([2.0, 1.0, 0.0])

def softmax(a):
    # Step 1: For numerical stability, subtract the max value from all logits
    #         Hint: shifted = a - np.max(a)
    shifted = a - np.max(a)

    # Step 2: Compute e^{a_k} for each k
    #         Hint: exp_a = np.exp(shifted)
    exp_a = np.exp(shifted)

    # Step 3: Divide each by the sum of all exponentials
    #         Hint: return exp_a / np.sum(exp_a)
    return exp_a / np.sum(exp_a)
    pass

probabilities = softmax(logits)
print(f"Logits:        {logits}")
print(f"Probabilities: {probabilities}")
print(f"Sum (should be 1.0): {np.sum(probabilities):.6f}")
```

**Bonus:** Why do we use softmax instead of simply dividing each logit by the sum of all logits?

Logits can be negative, and dividing a negative number by the sum would produce a negative probablity, which is meaningless. The exponential in softmax forces all values to be positive before normalizing, which guarantees a valid probability distribution.
